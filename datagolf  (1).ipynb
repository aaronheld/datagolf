```python
def winner_profile(event_id, year):
    linkbase = 'https://feeds.datagolf.com/historical-raw-data/rounds?tour=pga&year='
    link = linkbase + str(year) + '&event_id=' + str(event_id) + '&file_format=csv&key=5c80d222fec1f767e507b9cd7f69'
    tournament_results = pd.read_csv(link)
    results_nocut = tournament_results.loc[tournament_results['fin_text']!='CUT']
    results_nowd = results_nocut.loc[results_nocut['fin_text']!= 'WD']
    names = results_nowd['player_name']
    results = np.unique(results_nowd['fin_text'])
    cut_size = len(names) 
    weights = np.arange(1,cut_size+1)
    weights = weights/cut_size
    weights = reversed(weights)
    
    df = pd.DataFrame(columns = ['name','finish','sg_ott','sg_app','sg_putt','sg_arg'])
    df['name'] = np.unique(names)

    for name in names: 
        df.loc[df['name'] == name,'sg_ott'] = sum(results_nowd.loc[results_nowd['player_name'] == name]['sg_ott'])
        df.loc[df['name'] == name,'sg_app'] = sum(results_nowd.loc[results_nowd['player_name'] == name]['sg_app'])
        df.loc[df['name'] == name,'sg_putt'] = sum(results_nowd.loc[results_nowd['player_name'] == name]['sg_putt'])
        df.loc[df['name'] == name,'sg_arg'] = sum(results_nowd.loc[results_nowd['player_name'] == name]['sg_arg'])

    return df

    
    
winner_profile(7,2017)
```
